```python
%matplotlib inline
from matplotlib import pyplot as plt

import numpy as np
```

```python
log = np.load('../DeepDriving_TF/loss.npy')
loss = log[:,0]
time = log[:,1]

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(20,6))

ax[0].set_ylim(1e-5, 1e-0)
ax[0].set_xlim(-0.01 * time[-1],time[-1] * 1.01)
ax[0].semilogy(time[::10], loss[::10])
ax[0].set_xlabel('time, sec')
ax[0].set_ylabel('L2 loss')
# print(len(loss))
ax[0].set_title('MobileNet 80k iterations on GeForce 840M ')

log = np.load('../swap/loss.npy')
loss = log[:,0]
time = log[:,1]

ax[1].set_ylim(1e-5, 1e-0)
ax[1].set_xlim(-0.01 * time[-1],time[-1] * 1.01)
ax[1].semilogy(time[::10], loss[::10])
ax[1].set_xlabel('time, sec')
ax[1].set_ylabel('L2 loss')
#print(len(loss))
ax[1].set_title('AlexNet 140k iterations on GeForce GTX 1080')

plt.show()
```
